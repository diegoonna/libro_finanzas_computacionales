# Variables y Tipos de Datos Primitivos

En Python, **todo es un objeto**. Una variable es simplemente una etiqueta que apunta a un objeto en memoria. Python utiliza **tipado dinámico**: no necesitamos declarar el tipo de dato explícitamente.

## Tipos Numéricos

### Enteros (`int`)

Representan números del conjunto $\mathbb{Z}$. En Python 3, tienen precisión arbitraria.

**Aplicación Financiera:** Cantidades discretas como número de acciones, días al vencimiento, iteraciones de simulación.

```{shinylive-python}
#| label: tipos-int
# TODO: Definir variables enteras para un contexto financiero
# Ejemplo: cantidad de bonos, días al vencimiento, lotes de acciones
dias_vencimiento = 365
cantidad_bonos = 100
lotes = 10

print(f"Días al vencimiento: {dias_vencimiento} | Tipo: {type(dias_vencimiento)}")
```

### Punto Flotante (`float`)

Representan números reales mediante el estándar **IEEE 754**. Es el tipo estándar para precios, tasas y rendimientos.

::: {.callout-warning title="Error de Punto Flotante"}
Los números decimales como 0.1 no tienen representación binaria exacta. Para sistemas contables, usar la librería `decimal`.
:::

```{shinylive-python}
#| label: tipos-float
# Demostración del error de punto flotante
precio_a = 0.1
precio_b = 0.2
resultado = precio_a + precio_b

print(f"0.1 + 0.2 = {resultado}")
print(f"¿Es exactamente 0.3? {resultado == 0.3}")
```

### Booleanos (`bool`)

Valores lógicos `True` y `False`. Son la base de los algoritmos de decisión.

```{shinylive-python}
#| label: tipos-bool
mercado_abierto = True
volatilidad_alta = False

# True se comporta como 1, False como 0
señales_activas = True + True + False
print(f"Señales de compra activas: {señales_activas}")
```

### Cadenas de Texto (`str`)

Secuencias inmutables de caracteres. Esenciales para tickers, códigos ISIN y limpieza de datos.

```{shinylive-python}
#| label: tipos-str
ticker = "GGAL"
mercado = "BYMA"
instrumento = f"{ticker}.{mercado}"
print(f"Instrumento: {instrumento}")
```

# Conversión de Tipos (Casting)

Los datos del mundo real frecuentemente llegan como texto (desde CSVs o APIs). Es necesario convertirlos explícitamente.

```{shinylive-python}
#| label: casting
# Dato que llega como string desde una fuente externa
volumen_str = "5000"
precio_cierre = 150.50

# Conversión y operación
volumen = int(volumen_str)
monto_operado = volumen * precio_cierre

print(f"Monto operado: ${monto_operado:,.2f}")
```

---

# Operadores

## Operadores Aritméticos

```{shinylive-python}
#| label: operadores-aritmeticos
capital = 1_000_000
tasa = 0.08

# Operadores básicos
interes = capital * tasa
print(f"Interés simple: ${interes:,.2f}")

# Potencia (**) — útil para interés compuesto
periodos = 5
valor_futuro = capital * (1 + tasa) ** periodos
print(f"Valor futuro a {periodos} años: ${valor_futuro:,.2f}")
```

## Operadores de Comparación y Lógicos

| Operador | Significado |
|----------|-------------|
| `==`, `!=` | Igualdad / Desigualdad |
| `>`, `<`, `>=`, `<=` | Comparaciones |
| `and`, `or`, `not` | Lógica booleana |

```{shinylive-python}
#| label: operadores-logicos
precio = 150
media_movil = 140
rsi = 65

señal_compra = precio > media_movil and rsi < 70
print(f"¿Señal de compra? {señal_compra}")
```

---

# Entrada y Salida de Datos

## F-Strings (Formato Moderno)

```{shinylive-python}
#| label: f-strings
tasa = 0.05238
capital = 1_000_000
interes = capital * tasa

# Formatos: :.2% (porcentaje), :, (miles), :,.2f (decimal con miles)
print(f"Tasa: {tasa:.2%}")
print(f"Capital: ${capital:,}")
print(f"Interés generado: ${interes:,.2f}")
```
